uji t-test`

```python
#impoert library
from scipy import stats
```

```python
# Nilai yang ingin diuji (target 100)
nilai_target = 100
```

```python
# Melakukan uji t satu sampel
t_stat, p_value = stats.ttest_1samp(df1['TransactionAmount'], nilai_target)

# Output hasil uji t
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Tentukan hasil berdasarkan p-value
alpha = 0.05  # Tingkat signifikansi sebanyak 5%

if p_value < alpha:
    print("Tolak H₀: Ada perbedaan signifikan antara rata-rata sampel dan nilai target.")
else:
    print("Gagal tolak H₀: Tidak ada perbedaan signifikan antara rata-rata sampel dan nilai target.")
```
